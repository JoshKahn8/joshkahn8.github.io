```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv")
df.dropna()
df.head()
```

```python
chart1 = alt.Chart.from_dict({
  "data": {"url": "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv"
  },
  "mark": "bar",
  "height": 300,
  "width": 500,
  "encoding": {
    "x": {"field": "state", "type": "nominal"},
    "y": {"aggregate" : "count", "field": "state", "type": "quantitative"},
    "color": {"aggregate": "count", "type": "quantitative"} 
  }
})
chart1
```

For this visualization, I wanted to focus on displaying the total number of bigfoot sightings per state. My main goal was to be able to see which states have had the most bigfoot sightings and which states have had the least bigfoot sightings. According to my visualization, the state with the most bigfoot sightings is Washington with about 500 sightings. On the other hand, the states with the least amount of bigfoot sightings are Delaware and Rhode Island. Regarding the color scheme, the different shades of blue allow for the viewer to easily see which states have had the most bigfoot sightings based on how dark the shade of blue is. This detail isn't completely necessary as one can just simply look at which bar is the highest on the graph, but I felt that it makes the visualization more aestetichally pleasing. The only transformation that I performed on the data was to get rid of all of the NA values. The integrity of the data remained the same despite this transformation.

```python
bigfoot_url = 'https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/bfro_reports_fall2022.csv'
bigfoot = pd.read_csv(bigfoot_url)
```

```python
brush = alt.selection_interval(encodings=['x','y'])

dropdown = alt.binding_select(options=[None] + bigfoot['state'].unique().tolist(), 
                             labels = ['All']+ bigfoot['state'].unique().tolist())
selection_state = alt.selection_single(fields=['state'], bind=dropdown)


chart2 = alt.Chart(bigfoot_url).mark_rect().encode(
    alt.X("wind_bearing:Q", bin=alt.Bin(maxbins=50), title = 'Wind Bearing'),
    alt.Y("wind_speed:Q", bin=alt.Bin(maxbins=10), title='Wind Speed'),
    alt.Color("count()")
).properties(
    height=300
).transform_filter(
    selection_state
).add_selection(
    brush, selection_state
)

chart2
```

For this visualization, I once again wanted to do something with the bigfoot dataset but with an element of interactivity added in. I decided to compare the correlation between wind speed and wind bearing for each bigfoot sighting within the dataset. I chose to use bars for this visualization as I think that it is both the most visually pleasing and the most effective at displaying the correlation and reccurence of both variables.The element of interactivity that I added to this visualization is the ability for the user to choose which US state they specifically want to see the data for. Users will be able to use the first visualization to determine which state they want to research, allowing them to get the specific wind data that they need from a specific state. If I choose Washington, I can see that the most common wind speed and wind bearing correlation within the state is indicated by the dark blue rectangle. By looking at the color bar on the right I can see that the wind bearing and wind speed has been at that value for over 20 different bigfoot sightings within the dataset. If I was a conspiracy theorist or bigfoot hunter, I would be able to use this data to determine when I should go out to look for bigfoot in the state of Washington based on different attributes of the wind.
